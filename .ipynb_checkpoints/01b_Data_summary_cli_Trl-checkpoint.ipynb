```python


Ex_Ndata = Ex_data[['ID','Age_at_Surgery','Race','Surgery','Smoker','BMI',\
              'NAC','cT','pT','cT_or_pT','pN','Bx_Histology','Histology',\
              'Sample_weight_g_tumor','Tumor_digest_count_primary_tumor','Number_of_fragments_plated_tumor',\
              'Overall_TIL_growth']].copy()

# 'Surgeon'
Ex_Ndata = Ex_Ndata.replace(to_replace="Yes",value="Yes_TIL")
Ex_Ndata = Ex_Ndata.replace(to_replace="No",value="No_TIL")

Ex_Ndata = Ex_Ndata.replace(r'^\s*$', np.nan, regex=True) # Replace Blank values with DataFrame.replace() methods.

Extracted_col0 = Ex_Ndata.iloc[:,0:1]  # "ID"                        
Extracted_col2 = Ex_Ndata['Age_at_Surgery']
Extracted_col3 = Ex_Ndata['Race'];                              Extracted_col4 = Ex_Ndata['Surgery']
Extracted_col5 = Ex_Ndata['Smoker'];                            Extracted_col6 = Ex_Ndata['BMI']
Extracted_col7 = Ex_Ndata['NAC'];                               Extracted_col8 = Ex_Ndata['cT']
Extracted_col9 = Ex_Ndata['pT'];                                Extracted_col10 = Ex_Ndata['cT_or_pT']
Extracted_col11 = Ex_Ndata['pN'];                               Extracted_col12 = Ex_Ndata['Bx_Histology']
Extracted_col13 = Ex_Ndata['Histology'];                        Extracted_col14 = Ex_Ndata['Sample_weight_g_tumor']
Extracted_col15 = Ex_Ndata['Tumor_digest_count_primary_tumor']; Extracted_col16 = Ex_Ndata['Number_of_fragments_plated_tumor']
Extracted_col17 = Ex_Ndata['Overall_TIL_growth']

Ex_ndata_sel = Extracted_col0
# Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col1); 
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col2)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col3); Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col4)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col5); Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col6)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col7); Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col8)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col9); Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col10)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col11); Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col12)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col13); Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col14)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col15); Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col16)
Ex_ndata_sel = Ex_ndata_sel.join(Extracted_col17)

Ex_ndata_sel = Ex_ndata_sel.rename(columns={'Overall_TIL_growth': 'OverallTILGrowth'})

Cols = ['Age at Surgery','Race','Surgery','Smoker','BMI',\
        'NAC','cT','pT','cT or pT','pN','Bx Histology','Histology',\
        'Sample weight (g) tumor','Tumor digest count (primary tumor)','Number of fragments plated (tumor)',\
        'OverallTILGrowth']

feats = ['Age at Surgery','Race','Surgery','Smoker','BMI',\
        'NAC','cT','pT','cT or pT','pN','Bx Histology','Histology',\
        'Sample weight (g) tumor','Tumor digest count (primary tumor)','Number of fragments plated (tumor)']

feat_labels = feats
```
