```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import learning_curve, cross_val_score, StratifiedKFold
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, precision_score, recall_score

from scipy.optimize import curve_fit

import warnings
warnings.filterwarnings('ignore')

class LearningCurveAnalyzer:
    """
    A comprehensive class for generating and analyzing learning curves
    to assess dataset adequacy.
    """
    
    def __init__(self, X, y, task_type='classification'):
        """
        Initialize the analyzer.
        
        Parameters:
        -----------
        X : array-like, shape (n_samples, n_features)
            Feature matrix
        y : array-like, shape (n_samples,)
            Target variable
        task_type : str
            'classification' or 'regression'
        """
        self.X = X
        self.y = y
        self.task_type = task_type
        self.n_samples = X.shape[0]
        self.n_features = X.shape[1]
        
        print(f"Dataset Information:")
        print(f"  - Number of samples: {self.n_samples}")
        print(f"  - Number of features: {self.n_features}")
        print(f"  - Task type: {self.task_type}")
        if task_type == 'classification':
            unique, counts = np.unique(y, return_counts=True)
            print(f"  - Class distribution: {dict(zip(unique, counts))}")
        print()
    
    def generate_learning_curves(self, models=None, cv=9, n_jobs=-1):
        """
        Generate learning curves for multiple models.
        
        Parameters:
        -----------
        models : dict
            Dictionary of {model_name: model_object}
        cv : int
            Number of cross-validation folds
        n_jobs : int
            Number of parallel jobs
        
        Returns:
        --------
        results : dict
            Dictionary containing learning curve results for each model
        """
        if models is None:
            models = self._get_default_models()
        
        # Define training sizes (percentage of total training data)
        train_sizes = np.linspace(0.1, 1.0, 10)
        
        results = {}
        
        for model_name, model in models.items():
            print(f"Generating learning curve for {model_name}...")
            
            # Generate learning curve
            train_sizes_abs, train_scores, val_scores = learning_curve(
                model, self.X, self.y,
                train_sizes=train_sizes,
                cv=cv,
                scoring='accuracy' if self.task_type == 'classification' else 'r2',
                n_jobs=n_jobs,
                random_state=42,
                shuffle=True
            )
            
            results[model_name] = {
                'train_sizes': train_sizes_abs,
                'train_scores_mean': np.mean(train_scores, axis=1),
                'train_scores_std': np.std(train_scores, axis=1),
                'val_scores_mean': np.mean(val_scores, axis=1),
                'val_scores_std': np.std(val_scores, axis=1)
            }
        
        self.results = results
        return results
    
    def plot_learning_curves(self, figsize=(12, 10), save_path=None):
        """
        Plot learning curves for all models.
        
        Parameters:
        -----------
        figsize : tuple
            Figure size
        save_path : str
            Path to save the figure (optional)
        """
        if not hasattr(self, 'results'):
            raise ValueError("Must run generate_learning_curves() first")
        
        n_models = len(self.results)
        n_cols = 2
        n_rows = (n_models + 1) // 2
        
        fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
        axes = axes.flatten() if n_models > 1 else [axes]
        
        for idx, (model_name, result) in enumerate(self.results.items()):
            ax = axes[idx]
            
            # Plot training scores
            ax.plot(result['train_sizes'], result['train_scores_mean'],
                   'o-', color='#2ecc71', linewidth=2, markersize=6,
                   label='Training score')
            ax.fill_between(result['train_sizes'],
                           result['train_scores_mean'] - result['train_scores_std'],
                           result['train_scores_mean'] + result['train_scores_std'],
                           alpha=0.15, color='#2ecc71')
            
            # Plot validation scores
            ax.plot(result['train_sizes'], result['val_scores_mean'],
                   'o-', color='#e74c3c', linewidth=2, markersize=6,
                   label='Validation score')
            ax.fill_between(result['train_sizes'],
                           result['val_scores_mean'] - result['val_scores_std'],
                           result['val_scores_mean'] + result['val_scores_std'],
                           alpha=0.15, color='#e74c3c')
            
            # Calculate convergence metrics
            gap = result['train_scores_mean'][-1] - result['val_scores_mean'][-1]
            val_improvement = result['val_scores_mean'][-1] - result['val_scores_mean'][0]
            
            ax.set_xlabel('Sample Size', fontsize=12) #, fontweight='bold')
            ax.set_ylabel('Score', fontsize=12) #, fontweight='bold')
            ax.set_title(f'{model_name}') #\nGap: {gap:.3f} | Val Improvement: {val_improvement:.3f}',
                        #fontsize=12, fontweight='bold')
            ax.legend(loc='lower right', frameon=True, shadow=False)
            ax.grid(True, alpha=0.3)
            ax.set_ylim([0.40, 1.05])
        
        # Hide unused subplots
        for idx in range(n_models, len(axes)):
            axes[idx].set_visible(False)
        
        plt.suptitle('Learning Curves: ' f'{self.n_samples} Samples, {self.n_features} Features',
                    fontsize=16, fontweight='bold', y=0.995)
        plt.tight_layout()
        
        # Dataset Adequacy Assessment\n' +
        
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"Figure saved to {save_path}")
        
        return fig
    
    def analyze_adequacy(self):
        """
        Analyze whether the dataset is adequate based on learning curves.
        
        Returns:
        --------
        analysis : dict
            Dictionary containing adequacy analysis for each model
        """
        if not hasattr(self, 'results'):
            raise ValueError("Must run generate_learning_curves() first")
        
        analysis = {}
        
        print("=" * 70)
        print("DATASET ADEQUACY ANALYSIS")
        print("=" * 70)
        print()
        
        for model_name, result in self.results.items():
            # Calculate key metrics
            final_train_score = result['train_scores_mean'][-1]
            final_val_score = result['val_scores_mean'][-1]
            gap = final_train_score - final_val_score
            
            # Check if validation score is still improving
            val_scores = result['val_scores_mean']
            last_5_improvement = val_scores[-1] - val_scores[-5]
            
            # Determine adequacy
            is_converged = gap < 0.1 and last_5_improvement < 0.02
            is_adequate = final_val_score > 0.7 and is_converged
            
            # Recommendations
            if is_adequate:
                recommendation = "Dataset appears adequate for this model"
            elif gap > 0.15:
                recommendation = "High variance - consider regularization or more data"
            elif final_val_score < 0.7:
                recommendation = "Low performance - consider feature engineering or different model"
            elif not is_converged:
                recommendation = "Not converged - more data might improve performance"
            else:
                recommendation = "Dataset is likely adequate"
            
            analysis[model_name] = {
                'final_train_score': final_train_score,
                'final_val_score': final_val_score,
                'gap': gap,
                'last_5_improvement': last_5_improvement,
                'is_converged': is_converged,
                'is_adequate': is_adequate,
                'recommendation': recommendation
            }
            
            # Print analysis
            print(f"{model_name}:")
            print(f"  Training Score:     {final_train_score:.3f}")
            print(f"  Validation Score:   {final_val_score:.3f}")
            print(f"  Gap:                {gap:.3f}")
            print(f"  Recent Improvement: {last_5_improvement:.3f}")
            print(f"  {recommendation}")
            print()
        
        return analysis
    
    def plot_sample_efficiency(self, figsize=(12, 6), save_path=None):
        """
        Plot sample efficiency: validation score vs number of samples.
        
        This helps visualize diminishing returns and identify the optimal
        sample size for the task.
        """
        if not hasattr(self, 'results'):
            raise ValueError("Must run generate_learning_curves() first")
        
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=figsize)
        
        # Left plot: Validation score vs sample size
        for model_name, result in self.results.items():
            ax1.plot(result['train_sizes'], result['val_scores_mean'],
                    'o-', linewidth=2, markersize=6, label=model_name)
        
        ax1.axvline(x=self.n_samples, color='red', linestyle='--', 
                   linewidth=2, label='Current dataset size', alpha=0.7)
        ax1.set_xlabel('Number of Training Samples', fontsize=12, fontweight='bold')
        ax1.set_ylabel('Validation Score', fontsize=12, fontweight='bold')
        ax1.set_title('Sample Efficiency', fontsize=13, fontweight='bold')
        ax1.legend(loc='best', frameon=True, shadow=True)
        ax1.grid(True, alpha=0.3)
        
        # Right plot: Performance gain per additional sample
        for model_name, result in self.results.items():
            val_scores = result['val_scores_mean']
            train_sizes = result['train_sizes']
            
            # Calculate marginal improvement
            marginal_improvement = np.diff(val_scores) / np.diff(train_sizes)
            
            ax2.plot(train_sizes[1:], marginal_improvement,
                    'o-', linewidth=2, markersize=6, label=model_name)
        
        ax2.axhline(y=0, color='gray', linestyle='--', alpha=0.5)
        ax2.set_xlabel('Number of Training Samples', fontsize=12, fontweight='bold')
        ax2.set_ylabel('Marginal Performance Gain', fontsize=12, fontweight='bold')
        ax2.set_title('Diminishing Returns Analysis', fontsize=13, fontweight='bold')
        ax2.legend(loc='best', frameon=True, shadow=True)
        ax2.grid(True, alpha=0.3)
        
        plt.tight_layout()
        
        if save_path:
            plt.savefig(save_path, dpi=300, bbox_inches='tight')
            print(f"Figure saved to {save_path}")
        
        return fig
    
    def generate_summary_report(self):
        """
        Generate a comprehensive summary report.
        """
        if not hasattr(self, 'results'):
            raise ValueError("Must run generate_learning_curves() first")
        
        print("\n" + "=" * 70)
        print("COMPREHENSIVE SUMMARY REPORT")
        print("=" * 70)
        print()
        
        # Overall assessment
        analyses = self.analyze_adequacy()
        adequate_models = sum(1 for a in analyses.values() if a['is_adequate'])
        
        print(f"Overall Assessment:")
        print(f"  - {adequate_models}/{len(analyses)} models show adequate performance")
        print()
        
        # Best model
        best_model = max(analyses.items(), key=lambda x: x[1]['final_val_score'])
        print(f"Best Performing Model: {best_model[0]}")
        print(f"  - Validation Score: {best_model[1]['final_val_score']:.3f}")
        print()
        
        # Recommendations
        print("Recommendations:")
        if adequate_models >= len(analyses) / 2:
            print("  Your dataset of 131 samples appears adequate for this task")
            print("  The learning curves show convergence")
            print("  Consider this dataset size sufficient for deployment")
        else:
            print("  Consider collecting more data to improve performance")
            print("  Explore feature engineering to better utilize existing data")
            print("  Try ensemble methods or regularization techniques")
        
        print()
        print("=" * 70)
    
    def _get_default_models(self):
        """Get default models for analysis."""
        if self.task_type == 'classification':
            return {
                'Logistic Regression': Pipeline([
                    ('scaler', StandardScaler()),
                    ('clf', LogisticRegression(max_iter=1000, random_state=42))
                ]),
                'Random Forest': RandomForestClassifier(
                    n_estimators=100, random_state=42, max_depth=5
                ),
                'Gradient Boosting': GradientBoostingClassifier(
                    n_estimators=100, random_state=42, max_depth=3
                ),
                'RBF-SVM': Pipeline([
                    ('scaler', StandardScaler()),
                    ('clf', SVC(kernel='rbf', random_state=42))
                ])
            }
        else:
            from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
            from sklearn.svm import SVR
            from sklearn.linear_model import Ridge
            
            return {
                'Ridge Regression': Pipeline([
                    ('scaler', StandardScaler()),
                    ('reg', Ridge(random_state=42))
                ]),
                'Random Forest': RandomForestRegressor(
                    n_estimators=100, random_state=42, max_depth=5
                ),
                'Gradient Boosting': GradientBoostingRegressor(
                    n_estimators=100, random_state=42, max_depth=3
                ),
                'SVR': Pipeline([
                    ('scaler', StandardScaler()),
                    ('reg', SVR(kernel='rbf'))
                ])
            }
        


    def estimate_required_samples(self, target_fraction=0.95):
        """
        Estimate how many samples would be required if learning curve
        has not converged by fitting a saturating function.

        Parameters:
        -----------
        target_fraction : float
            Fraction of asymptotic performance to reach (default=0.95)

        Returns:
        --------
        estimates : dict
        """
        if not hasattr(self, 'results'):
            raise ValueError("Must run generate_learning_curves() first")

        def saturating_func(x, a, b, c):
            # a = asymptote
            # b = growth rate
            # c = horizontal shift
            return a * (1 - np.exp(-b * (x - c)))

        estimates = {}

        print("\n" + "="*70)
        print("SAMPLE SIZE EXTRAPOLATION ANALYSIS")
        print("="*70)
        print()

        for model_name, result in self.results.items():
            x = result['train_sizes']
            y = result['val_scores_mean']

            try:
                popt, _ = curve_fit(
                    saturating_func,
                    x,
                    y,
                    bounds=([0.5, 1e-6, 0], [1.5, 1, np.max(x)]),
                    maxfev=10000
                )

                asymptote = popt[0]
                target_score = target_fraction * asymptote

                # Solve for required x
                required_samples = (
                    -np.log(1 - target_score / asymptote) / popt[1]
                ) + popt[2]

                required_samples = int(np.ceil(required_samples))

                converged = result['val_scores_mean'][-1] >= target_score

                if converged:
                    msg = "Already near asymptotic performance"
                else:
                    additional_needed = required_samples - self.n_samples
                    msg = f"Approx. {max(additional_needed,0)} more samples needed"

                print(f"{model_name}:")
                print(f"  Estimated asymptotic score: {asymptote:.3f}")
                print(f"  Target (95%): {target_score:.3f}")
                print(f"  Required samples: {required_samples}")
                print(f"  {msg}")
                print()

                estimates[model_name] = {
                    "asymptote": asymptote,
                    "target_score": target_score,
                    "required_samples": required_samples,
                    "additional_needed": max(required_samples - self.n_samples, 0)
                }

            except:
                print(f"{model_name}: Could not reliably extrapolate.\n")

        return estimates

    def plot_powerlaw_extrapolation(self, target_fraction=0.95, figsize=(12, 10)):
        """
        Plot power-law extrapolation against observed learning curves.
        """
        from scipy.optimize import curve_fit

        if not hasattr(self, 'results'):
            raise ValueError("Must run generate_learning_curves() first")

        def power_law(x, a, b, c):
            # a = asymptote
            # b = scale
            # c = decay rate
            return a - b * (x ** (-c))

        n_models = len(self.results)
        n_cols = 2
        n_rows = (n_models + 1) // 2

        fig, axes = plt.subplots(n_rows, n_cols, figsize=figsize)
        axes = axes.flatten() if n_models > 1 else [axes]

        for idx, (model_name, result) in enumerate(self.results.items()):
            ax = axes[idx]

            x = result['train_sizes']
            y = result['val_scores_mean']

            # Fit power-law model
            try:
                popt, _ = curve_fit(
                    power_law,
                    x,
                    y,
                    bounds=([0.4, 0, 0], [1.2, 10, 5]),
                    maxfev=10000
                )

                asymptote = popt[0]
                target_score = target_fraction * asymptote

                # Extended sample range for projection
                x_extended = np.linspace(min(x), self.n_samples * 3, 200)
                y_extended = power_law(x_extended, *popt)

                # Plot observed data
                ax.plot(x, y, 'o', label='Observed validation', markersize=6)

                # Plot fitted curve
                ax.plot(x_extended, y_extended, '-', linewidth=2,
                        label='Power-law extrapolation')
                ax.set_xlim(0, 100)
                ax.set_ylim(0.4, 1.0)

                # # Plot asymptote and 95% line
                # ax.axhline(asymptote, linestyle='--', alpha=0.7,
                #            label=f'Asymptote ({asymptote:.3f})')
                # ax.axhline(target_score, linestyle=':', alpha=0.7,
                #            label=f'95% target ({target_score:.3f})')

                ax.set_title(model_name)
                ax.set_xlabel("Number of Training Samples")
                ax.set_ylabel("Validation Score")
                ax.grid(True, alpha=0.3)
                ax.legend()

            except:
                ax.text(0.5, 0.5, "Fit Failed", ha='center')
                ax.set_title(model_name)

        for idx in range(n_models, len(axes)):
            axes[idx].set_visible(False)

        plt.tight_layout()
        return fig

```
